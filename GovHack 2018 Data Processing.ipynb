```python
import requests
import json
import pandas as pd
```

```python
res = requests.get("https://www.accesscanberra.act.gov.au/cc/mappingService/getDataPotholes")
potholes = res.json()
```

```python
res = requests.get("https://www.accesscanberra.act.gov.au/cc/mappingService/getDataTrollies")
trollies = res.json()
```

```python
res = requests.get("https://www.accesscanberra.act.gov.au/cc/mappingService/getDataStreetlights")
streetlights = res.json()
```

```python
res = requests.get("https://www.accesscanberra.act.gov.au/cc/mappingService/getDataTrees")
trees = res.json()
```

```python
potholes_df = pd.read_json(potholes)
trollies_df = pd.read_json(trollies)
streetlights_df = pd.read_json(streetlights)
trees_df = pd.read_json(trees)
df = potholes_df.append(trollies_df)
df = df.append(streetlights_df)
df = df.append(trees_df)
```

```python
drink_df = pd.read_csv('../ACT Location Data/Drinking_Fountains.csv')
drink_df.rename(columns = { 'LOCATION':'LOCATION_DESCRIPTION', 'category':'CATEGORY' }, inplace = True)
drink_df['CATEGORY'] = 'Drinking Fountain'
drink_df['TYPE'] = 'Drinking Fountain'
drink_df.head()
```

```python
toilet_df = pd.read_csv('../ACT Location Data/Public_Toilets_in_the_ACT.csv')
toilet_df.rename(columns = {'ASSET_ID':'ID','DIVISION_NAME':'DIVISION', 'TOILET_TYPE_TEXT':'TYPE', 'Location':'Location_1'}, inplace = True)
toilet_df['TYPE'] = toilet_df.TYPE.str.title()
toilet_df['LOCATION_DESCRIPTION'] = toilet_df.LOCATION_DESCRIPTION.str.title()
toilet_df['DIVISION'] = toilet_df.DIVISION.str.title()
toilet_df['CATEGORY'] = 'Toilet'
toilet_df.head()
```

```python
furn_df = pd.read_csv('../ACT Location Data/Public_Furniture_in_the_ACT.csv')
furn_df.rename(columns = {'ASSET ID':'ID', 'FEATURE TYPE':'TYPE', 'LOCATION NAME':'LOCATION_DESCRIPTION', 'DIVISION NAME':'DIVISION', 'TOILET_TYPE_TEXT':'TYPE', 'Location':'Location_1'}, inplace = True)
furn_df['TYPE'] = furn_df.TYPE.str.title()
furn_df['LOCATION_DESCRIPTION'] = furn_df.LOCATION_DESCRIPTION.str.title()
furn_df['DIVISION'] = furn_df.DIVISION.str.title()
furn_df['CATEGORY'] = 'Furniture'
furn_df.head()
```

```python
amen_df = drink_df.append(toilet_df).append(furn_df)
amen_df = amen_df[['ID','CATEGORY','DIVISION','LATITUDE','LONGITUDE','LOCATION_DESCRIPTION','TYPE']]
amen_df.sample(20)
```

```python
amen_df.to_csv('ACT_Amenities.csv', index=False)
```

```python
bbq_df = pd.read_csv('../ACT Location Data/Public_Barbeques_in_the_ACT.csv')
bbq_df.rename(columns = {'ASSET_ID':'ID', 'BBQ_TYPE':'TYPE', 'LOCATION':'DIVISION'}, inplace = True)
bbq_df['CATEGORY'] = 'BBQ'
bbq_df.head()
```

```python
pg_df = pd.read_csv('../ACT Location Data/Town_And_District_Playgrounds.csv')
pg_df.rename(columns = {'LOCATION':'LOCATION_DESCRIPTION', 'ASSET_ID':'ID', 'EQUIPMENT':'TYPE', 'DIVISION_NAME':'DIVISION'}, inplace = True)
pg_df['CATEGORY'] = 'Playground'
pg_df.head()
```

```python
fam_df = bbq_df.append(pg_df)
fam_df = fam_df[['ID','CATEGORY','DIVISION','LATITUDE','LONGITUDE','LOCATION_DESCRIPTION','TYPE']]
fam_df.sample(10)
```

```python
fam_df.to_csv('ACT_Family_Places.csv', index=False)
```
